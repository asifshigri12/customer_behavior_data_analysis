```python
# Create a column age_group
labels =['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
from sqlalchemy.types import String, Integer, Float

# Step 1: Fix column names for MySQL (no spaces or special characters)
df.columns = [c.replace(' ', '_').replace('-', '_') for c in df.columns]

# Step 2: Map column data types
dtype_map = {}
for col in df.columns:
    if df[col].dtype == 'int64':
        dtype_map[col] = Integer()
    elif df[col].dtype == 'float64':
        dtype_map[col] = Float()
    else:
        dtype_map[col] = String(255)

# Step 3: Save DataFrame to MySQL
df.to_sql(
    name="customer_analysis",  # table name
    con=engine,
    if_exists="replace",       # overwrite table if exists
    index=False,
    dtype=dtype_map
)

print("DataFrame saved to MySQL successfully âœ…")

```
